```python
# import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv_path = "./archive/birds.csv"

# create pandas Dataframe
df = pd.read_csv(csv_path)
df.head()
```

```python
# see filepaths formats
df["filepaths"].apply(lambda x: x.split(".")[-1]).value_counts()
```

```python
# How much different Labels we have (should be 525)
len(df["labels"].value_counts())
```

```python
# print 10 example labels
df["labels"].value_counts()[:10]
```

```python
df["data set"].value_counts().plot(
    kind="bar",
    figsize=(5, 5),
    title="Images per Directory",
    ylabel="Images",
    xlabel="Directories",
)
plt.show()
```

```python
# show the image balance in our train directory
df.query("`data set` == 'train'")["labels"].value_counts().plot(kind="barh")
plt.axis("off")
plt.show()
```

```python
# show the image balance in our test directory
df.query("`data set` == 'test'")["labels"].value_counts().plot(kind="barh")
plt.axis("off")
plt.show()
```

```python
# show the image balance in our valid directory
df.query("`data set` == 'valid'")["labels"].value_counts().plot(kind="barh")
plt.axis("off")
plt.show()
```
